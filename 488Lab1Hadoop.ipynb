```python
%%writefile RatingBreakdown.py
# import modules
from mrjob.job import MRJob
from mrjob.step import MRStep

# create class inhereted from MRJob
class RatingBreakdown(MRJob):
  # assign steps, first mapper last reducer
  def steps(self):
    return [
            MRStep(mapper=self.mapper_get_rating,
                   reducer=self.reducer_count_ratings)
    ]

  # creating mapper, assigning attributes from dataset
  def mapper_get_rating(self, _, line):
    (userID, movieID, rating, timestamp) = line.split('\t')
    # assign like the key rating and assign each row value 1
    yield rating, 1

  # creating reducer, sum
  def reducer_count_ratings(self, key, values):
    # in function of each key we sum values
    yield key, sum(values)

if __name__ == '__main__':
  RatingBreakdown.run()
```

```python
%%writefile MovieAvgRating.py
# import modules
from mrjob.job import MRJob
from mrjob.step import MRStep

# create class inherited from MRJob
class MovieAvgRating(MRJob):
    # assign steps, first mapper last reducer
    def steps(self):
        return [
            MRStep(mapper=self.mapper,
                   reducer=self.reducer)
        ]

    # creating mapper, assigning attributes from dataset
    def mapper(self, _, line):
        # Expecting data in the format: userID\tmovieID\trating\ttimestamp
        userID, movieID, rating, timestamp = line.split('\t')
        # yield movieID as the key and rating as the value
        yield movieID, float(rating)

    # creating reducer to calculate average rating
    def reducer(self, movieID, ratings):
        # Calculate the average rating
        ratings_list = list(ratings)
        average_rating = sum(ratings_list) / len(ratings_list)
        yield movieID, average_rating

if __name__ == '__main__':
    MovieAvgRating.run()
```

```python
%%writefile MovieMaxRating.py
# import modules
from mrjob.job import MRJob
from mrjob.step import MRStep

# create class inherited from MRJob
class MovieMaxRating(MRJob):
  # assign steps, first mapper last reducer
  def steps(self):
    return [
            MRStep(mapper=self.mapper,
                   reducer=self.reducer),
            MRStep(reducer = self.reducer2)
    ]

  # creating mapper, assigning attributes from dataset
  def mapper(self, _, line):
    (userID, movieID, rating, timestamp) = line.split('\t')
    # assign like the key rating and assign each row value 1
    yield movieID, rating

  # creating reducer, sum
  def reducer(self, key, values):
    # in function of each key we sum values
    sum = 0
    count = 0
    for v in values:
        sum += float(v)
        count += 1
    yield None, ((sum/count), key)

  def reducer2 (self, key, values):
    yield 'Max:', max(values)

if __name__ == '__main__':
  MovieMaxRating.run()
```

```python
%%writefile TopKey.py
# import modules
from mrjob.job import MRJob
from mrjob.step import MRStep

# create class inherited from MRJob
class TopKey(MRJob):
  # assign steps, first mapper last reducer
  def steps(self):
    return [
            MRStep(mapper=self.mapper,
                   reducer=self.reducer),
            MRStep(reducer=self.reducer2)
    ]

  # creating mapper, assigning attributes from dataset
  def mapper(self, _, line):
    (userID, movieID, rating, timestamp) = line.split('\t')
    # assign movieID as the key and rating as the value
    yield movieID, float(rating)

  # creating reducer to calculate average rating
  def reducer(self, key, values):
    # calculate the sum and count for each key
    sum_ratings = 0
    count = 0
    for v in values:
        sum_ratings += v
        count += 1
    # yield None as the key to combine all movies in the final reducer
    yield None, (sum_ratings / count, key)

  # reducer to find the top 3 movies by average rating
  def reducer2(self, _, values):
    # sort values in descending order by average rating and take the top 3
    top_3 = sorted(values, reverse=True)[:3]
    for avg_rating, movieID in top_3:
      yield movieID, avg_rating

if __name__ == '__main__':
  TopKey.run()
```

```python
%%writefile UserProfessionCount.py
# Import modules
from mrjob.job import MRJob
from mrjob.step import MRStep


class UserProfessionCount(MRJob):

    def steps(self):
        return [
            MRStep(mapper=self.mapper_get_profession,
                   reducer=self.reducer_count_professions)
        ]


    def mapper_get_profession(self, _, line):
        fields = line.split('|')
        if len(fields) >= 4:
            profession = fields[3]
            yield profession, 1


    def reducer_count_professions(self, key, values):
        yield key, sum(values)

if __name__ == '__main__':
    UserProfessionCount.run()

```

```python
%%writefile AverageAgeByGender.py
# Import modules
from mrjob.job import MRJob
from mrjob.step import MRStep


class AverageAgeByGender(MRJob):

    def steps(self):
        return [
            MRStep(mapper=self.mapper_get_age_gender,
                   reducer=self.reducer_average_age)
        ]

    def mapper_get_age_gender(self, _, line):
        fields = line.split('|')
        if len(fields) >= 4:
            age = int(fields[1])
            gender = fields[2]
            yield gender, age

    def reducer_average_age(self, key, values):
        total_age = 0
        count = 0
        for age in values:
            total_age += age
            count += 1
        average_age = total_age / count if count > 0 else 0
        yield key, average_age

if __name__ == '__main__':
    AverageAgeByGender.run()

```

```python
%%writefile OldestUserByGender.py
# Import modules
from mrjob.job import MRJob
from mrjob.step import MRStep

class OldestUserByGender(MRJob):

    def steps(self):
        return [
            MRStep(mapper=self.mapper_get_age_gender,
                   reducer=self.reducer_find_oldest)
        ]

    def mapper_get_age_gender(self, _, line):
        fields = line.split('|')
        if len(fields) >= 4:
            age = int(fields[1])
            gender = fields[2]
            yield gender, age

    def reducer_find_oldest(self, key, values):
        oldest_age = -1
        for age in values:
            if age > oldest_age:
                oldest_age = age
        yield key, oldest_age

if __name__ == '__main__':
    OldestUserByGender.run()
```
