```python
%%writefile Top10MostPurchasedItems.py
from mrjob.job import MRJob
from mrjob.step import MRStep

class Top10MostPurchasedItems(MRJob):
    def mapper(self, _, line):
        # Initialize a static counter for the row number in the mapper function
        if not hasattr(self, 'row_count'):
            self.row_count = 0

        # Skip the first row (header)
        if self.row_count == 0:
            self.row_count += 1
            return
        else:
            self.row_count += 1

        # Split each line by commas to process CSV format
        values = line.strip().split(',')

        # Assuming the first 3 columns are Invoice, Price, and Quantity and the last column is Country
        # We need to extract the item columns between the 4th column (index 3) and the 678th column (index 678)
        items = values[3:678]  # Adjusted to exclude non-item columns

        # Yield each item as a key with a count of 1
        for item in items:
            item = item.strip()
            if item:  # Only yield non-empty items
                yield item, 1

    def reducer(self, key, counts):
        # Sum all occurrences of each unique item
        yield None, (sum(counts), key)

    def reducer_find_top_10(self, _, item_count_pairs):
        # Sort all items by count in descending order and take the top 10
        top_10 = sorted(item_count_pairs, reverse=True, key=lambda x: x[0])[:10]
        for count, item in top_10:
            yield item, count

    def steps(self):
        return [
            MRStep(mapper=self.mapper, reducer=self.reducer),
            MRStep(reducer=self.reducer_find_top_10)
        ]

if __name__ == '__main__':
    Top10MostPurchasedItems.run()


```

```python
%%writefile TransactionsByCountry.py
from mrjob.job import MRJob
from mrjob.step import MRStep

class TransactionsByCountry(MRJob):
    def mapper(self, _, line):
        # Initialize a static counter for the row number in the mapper function
        if not hasattr(self, 'row_count'):
            self.row_count = 0

        # Skip the first row (header)
        if self.row_count == 0:
            self.row_count += 1
            return
        else:
            self.row_count += 1

        # Split each line by commas to process CSV format
        values = line.strip().split(',')

        # Assuming 'Country' is in the last column, adjust index if needed
        country = values[-1].strip()  # Get the country value
        if country:  # Yield country if it is not empty
            yield country, 1

    def reducer(self, key, counts):
        # Sum all occurrences of each country
        yield None, (sum(counts), key)

    def reducer_find_top_countries(self, _, country_count_pairs):
        # Sort all countries by transaction count in descending order and take top countries
        top_countries = sorted(country_count_pairs, reverse=True, key=lambda x: x[0])
        for count, country in top_countries:
            yield country, count

    def steps(self):
        return [
            MRStep(mapper=self.mapper, reducer=self.reducer),
            MRStep(reducer=self.reducer_find_top_countries)
        ]

if __name__ == '__main__':
    TransactionsByCountry.run()
```

```python
%%writefile MostSellingItemCounter.py
from mrjob.job import MRJob
from mrjob.step import MRStep

class MostSellingItemCounter(MRJob):
    def mapper(self, _, line):
        # Initialize a static counter for the row number in the mapper function
        if not hasattr(self, 'row_count'):
            self.row_count = 0

        # Skip the first row (header)
        if self.row_count == 0:
            self.row_count += 1
            return
        else:
            self.row_count += 1

        # Split each line by commas to process CSV format
        values = line.strip().split(',')

        # Skipping the first 3 columns: Invoice, Price, Quantity, and the last column: Country
        # Assuming items start from the 4th column (index 3) and go up to the second last column
        items = values[3:-1]  # Exclude Invoice (0), Price (1), Quantity (2), and Country (last column)

        # Yield each item as a key with a count of 1
        for item in items:
            item = item.strip()
            if item:  # Only yield non-empty items
                yield item, 1

    def reducer(self, key, counts):
        # Sum all occurrences of each unique item
        yield None, (sum(counts), key)

    def reducer_find_most_selling(self, _, item_count_pairs):
        # Find the item with the maximum count
        yield max(item_count_pairs)

    def steps(self):
        return [
            MRStep(mapper=self.mapper, reducer=self.reducer),
            MRStep(reducer=self.reducer_find_most_selling)
        ]

if __name__ == '__main__':
    MostSellingItemCounter.run()
```

```python
%%writefile ItemsFrequentlyBoughtTogether.py
from mrjob.job import MRJob
from mrjob.step import MRStep
from itertools import combinations

class ItemsFrequentlyBoughtTogether(MRJob):
    def mapper(self, _, line):
        # Initialize a static counter for the row number in the mapper function
        if not hasattr(self, 'row_count'):
            self.row_count = 0

        # Skip the first row (header)
        if self.row_count == 0:
            self.row_count += 1
            return
        else:
            self.row_count += 1

        # Split each line by commas to process CSV format
        values = line.strip().split(',')

        # Extract items from the transaction (ignoring transaction ID or other columns)
        items = [item.strip() for item in values[1:] if item.strip()]

        # Generate all combinations of item pairs within a single transaction
        for pair in combinations(items, 2):  # Generate pairs of items
            yield tuple(sorted(pair)), 1  # Sort the pair for consistent output

    def reducer(self, key, counts):
        # Sum all occurrences of each pair
        yield key, sum(counts)

    def steps(self):
        return [
            MRStep(mapper=self.mapper, reducer=self.reducer)
        ]

if __name__ == '__main__':
    ItemsFrequentlyBoughtTogether.run()
```

```python
%%writefile Top5ItemPairs.py
from mrjob.job import MRJob
from mrjob.step import MRStep
from itertools import combinations

class Top5ItemPairs(MRJob):
    def mapper(self, _, line):
        # Initialize a static counter for the row number in the mapper function
        if not hasattr(self, 'row_count'):
            self.row_count = 0

        # Skip the first row (header)
        if self.row_count == 0:
            self.row_count += 1
            return
        else:
            self.row_count += 1

        #
        values = line.strip().split(',')


        items = values[3:-1]


        for pair in combinations(items, 2):
            yield tuple(sorted(pair)), 1  # Sort the pair for consistent output

    def reducer_count_pairs(self, pair, counts):
        # Sum all occurrences of each pair
        yield None, (sum(counts), pair)

    def reducer_find_top_5(self, _, pair_count_pairs):
        # Find the top 5 most frequent pairs
        top_5 = sorted(pair_count_pairs, reverse=True, key=lambda x: x[0])[:5]
        for count, pair in top_5:
            yield pair, count

    def steps(self):
        return [
            MRStep(mapper=self.mapper, reducer=self.reducer_count_pairs),
            MRStep(reducer=self.reducer_find_top_5)
        ]

if __name__ == '__main__':
    Top5ItemPairs.run()
```

```python
%%writefile AverageQuantityPerTransaction.py
from mrjob.job import MRJob
from mrjob.step import MRStep

class AverageQuantityPerTransaction(MRJob):
    def mapper(self, _, line):
        # Initialize a static counter for the row number in the mapper function
        if not hasattr(self, 'row_count'):
            self.row_count = 0

        # Skip the first row (header)
        if self.row_count == 0:
            self.row_count += 1
            return
        else:
            self.row_count += 1

        # Split each line by commas to process CSV format
        values = line.strip().split(',')

        # Assuming the first column is TransactionID, and subsequent columns are quantities sold
        transaction_id = values[0].strip()
        quantities = [int(q.strip()) for q in values[1:] if q.strip().isdigit()]

        # Emit the transaction ID and the total quantity for that transaction
        yield transaction_id, sum(quantities)

    def reducer_sum_transactions(self, transaction_id, quantities):
        # Emit each transaction's total quantity sold
        yield None, sum(quantities)

    def reducer_calculate_average(self, _, total_quantities):
        # Calculate the average quantity per transaction
        total_quantity = 0
        transaction_count = 0

        for quantity in total_quantities:
            total_quantity += quantity
            transaction_count += 1

        # Emit the average
        yield "Average Quantity Sold per Transaction", total_quantity / transaction_count

    def steps(self):
        return [
            MRStep(mapper=self.mapper, reducer=self.reducer_sum_transactions),
            MRStep(reducer=self.reducer_calculate_average)
        ]

if __name__ == '__main__':
    AverageQuantityPerTransaction.run()
```

```python
%%writefile AssociationRuleMining.py
from mrjob.job import MRJob
from mrjob.step import MRStep
from itertools import combinations

class AssociationRuleMining(MRJob):

    def configure_args(self):
        super(AssociationRuleMining, self).configure_args()
        # Minimum support and confidence can be passed as arguments
        self.add_passthru_arg('--min_support', type=int, default=10, help='Minimum support threshold')
        self.add_passthru_arg('--min_confidence', type=float, default=0.8, help='Minimum confidence threshold')

    def mapper_find_frequent_itemsets(self, _, line):
        # Skip header row
        if line.startswith('TransactionID'):
            return

        # Split the line by commas to extract items
        values = line.strip().split(',')
        items = [item.strip() for item in values[1:] if item.strip()]

        # Emit all itemsets (combinations) of size 1 to n
        for i in range(1, len(items) + 1):
            for itemset in combinations(items, i):
                yield itemset, 1

    def reducer_count_itemsets(self, itemset, counts):
        # Aggregate counts for each itemset
        total_count = sum(counts)
        if total_count >= self.options.min_support:
            yield itemset, total_count

    def mapper_generate_rules(self, itemset, count):
        # Generate association rules from frequent itemsets
        itemset = list(itemset)
        if len(itemset) > 1:
            for i in range(1, len(itemset)):
                for antecedent in combinations(itemset, i):
                    consequent = tuple(set(itemset) - set(antecedent))
                    yield (antecedent, consequent), count

    def reducer_calculate_confidence(self, rule, counts):
        # Calculate confidence for each rule
        total_count = sum(counts)
        antecedent, consequent = rule

        # Reuse map-reduce to calculate the antecedent count
        yield antecedent, (consequent, total_count)

    def reducer_filter_rules(self, antecedent, consequent_counts):
        # Calculate confidence for each rule and filter based on min confidence
        antecedent_count = sum([count for consequent, count in consequent_counts])
        for consequent, rule_count in consequent_counts:
            confidence = rule_count / antecedent_count
            if confidence >= self.options.min_confidence:
                yield (antecedent, consequent), confidence

    def steps(self):
        return [
            MRStep(mapper=self.mapper_find_frequent_itemsets,
                   reducer=self.reducer_count_itemsets),
            MRStep(mapper=self.mapper_generate_rules,
                   reducer=self.reducer_calculate_confidence),
            MRStep(reducer=self.reducer_filter_rules)
        ]

if __name__ == '__main__':
    AssociationRuleMining.run()
```
